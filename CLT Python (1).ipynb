# The Central Limit Theorem (CLT)

Basically, the Central Limit Theorem states that the distribution of the sum (or average) of a large number of i.i.d. random variables converges in law to a Normal distribution regardless of the original distribution.

Let $X_{1}, X_{2}, ... , X_{n}$ be a sequence of independent and identically distributed ($i.i.d$) random variables with $E[X] = \mu$, $< \infty$ (finite first moment), and $Var(X) = \sigma^{2}$ exists.

\vspace{10}
\noindent
\textbf{(CLT)} As $n$ becomes large, then the distribution of $\frac{1}{n} \sum_{i=1}^{n} x_{i}$ (the sample mean) is approximately a Normal distribution $N \bigg( \mu, \frac{\sigma^{2}}{n} \bigg)$. 


## Useful direct consequence

As a consequence, we also have that

$$
\frac{\sum_{i=1}^{n} (x_{i} - \mu)}{\sqrt{n} \  \sigma} \xrightarrow{ \ \ L \ \ } N(0,1) \\
$$

Indeed, we have that

$$
\begin{align*}
& \sum_{i=1}^{n} (x_{i} - \mu) \  \xrightarrow{ \ \ L \ \ } N(0, n   \sigma^{2}) \\
& \sum_{i=1}^{n} x_{i} \  \xrightarrow{ \ \ L \ \ } N( n  \mu , n  \sigma^{2}) \\
&\frac{1}{n}  \sum_{i=1}^{n} x_{i} \  \xrightarrow{ \ \ L \ \ } N \bigg( \mu , \frac{ \sigma^{2}}{n} \bigg) \\
\end{align*}
$$

## Example 1: Exponential variables


### Example 1

Let $y_{1}, y_{2},..., y_{n}$ be $i.i.d$ exponential realizations of  $Y_{i} \sim \mathcal{E}(\lambda = 5)$ and $n=1,000$. The theoretical mean is $E[Y] = 1/ \lambda = 1/5 = 0.2$ and the theoretical variance is $Var(Y) = 1/5^{2}$. Then we expect the sample average $ \frac{1}{n
} \sum_{i=1}^{n} y_{i}$ to converge to a {\color{red}Normal distribution}
 $N\bigg(  E[Y] = \mu=\frac{1}{5},   \frac{\sigma^{2}}{n}= \frac{1}{25,000}   \bigg)

In other words: The distribution of the sample means of Exponentially distributed samples will obey a Normal distribution.



```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

np.random.seed(2023)
lambda_val = 5
n = 1000
x = np.random.exponential(scale=1/lambda_val, size=n)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(x, bins=30, density=True, color='gray', alpha=0.8)
plt.plot(np.linspace(0, 2, 100), lambda_val * np.exp(-lambda_val * np.linspace(0, 2, 100)), 'r', linewidth=2)
plt.xlabel('Y'); plt.ylabel('Density')
plt.title('Distribution of Exponential random variates')
plt.text(0.75, 2, r'$E[Y] = \frac{1}{5}$', fontsize=12)
plt.text(0.75, 1, r'$Var[Y] = \frac{1}{25}$', fontsize=12)
plt.grid()

sample_means = np.zeros(100000)
for i in range(100000):
    sample_means[i] = np.mean(np.random.exponential(scale=1/lambda_val, size=n))

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True, color='gray', alpha=0.8)
xs = np.linspace(0.17, 0.23, 100)
pdf = norm.pdf(xs, 0.2, 25/4000)
plt.plot(xs, pdf, 'k', linewidth=2, color = 'red')
plt.xlabel('W'); plt.ylabel('Density')
plt.title('Distribution of the empirical mean')
plt.text(0.18, 40, r'$E[W] = \frac{1}{5}$', fontsize=12)
plt.text(0.215, 40, r'$Var[W] = \frac{1}{25000}$', fontsize=12)
plt.grid()

plt.tight_layout()
plt.show()
```

## Example 2: Application 1/3

### Example 2 
According to a certain maritime organization, the distribution of the length of a specific fish species at maturity in the Mekong delta has mean $\mu= 83$ cm. and standard deviation $\sigma = 7$ cm. The random variable $Y$ denotes the length of these fishes.

(i) Suppose we sample $25$ individuals. What is the probability that the sample average is above $86$ cm. ?

(i) From the CLT, we know that the distribution of the sample average  $ \frac{1}{25} \sum_{i=1}^{25} y_{i}$ has a  {\color{red}Normal distribution}
\\ $N\bigg(  \mu = 83, \frac{\sigma}{\sqrt{n}}= \frac{7}{5}    \bigg)$.

 So, $P\bigg(  \overline{y} > 86  \bigg) =  P\bigg(  Z > \frac{\overline{y}-\mu}{ \sigma / \sqrt{n}}  \bigg) = P\bigg(  Z > \frac{86-83}{1.4}  \bigg)  = P \bigg(  Z > 2.142857  \bigg)  = \mathbf{0.01606 \approx 1.6 \%}$


```python
1 - norm.cdf(86, loc=83, scale=7/5)
```
